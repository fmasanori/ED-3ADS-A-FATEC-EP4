```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
low_memory=False
%matplotlib inline

print ('An√°lise de Dados - EP4 - 3 A ADS - Estrutura de Dados')

arquivo = 'CovidSP.csv'
dados = pd.read_csv(arquivo)
dados.rename(columns={'epidemiological_week': 'SEMANA',
                   'date': 'DATA',
                   'order_for_place': 'ORDEM',
                   'state': 'UF',
                   'city': 'CIDADE',
                   'city_ibge_code': 'CODIGO',
                   'place_type': 'TIPO',
                   'last_available_confirmed': 'TOTALCONFIRMADOS',
                   'last_available_confirmed_per_100k_inhabitants': 'ULTIMOPOR100K',
                   'new_confirmed': 'NOVOSCONFIRMADOS',
                   'last_available_deaths': 'ULTIMOSOBITOS',
                   'new_deaths': 'NOVOSOBITOS',
                   'last_available_death_rate': 'ULTIMORATEOBITOS',
                   'estimated_population_2019': 'POPULACAO',
                   'is_last': 'ULTIMO',
                   'is_repeated': 'REPETIDO'
                  }, inplace=True)

dados.groupby('CIDADE').size().sort_values().tail(50).plot(kind='barh', figsize=(10,10), title='50 cidades com mais casos de COVID-19 em SP')
```
